```python
df = pd.read_csv(r"C:\Users\laksh\Downloads\NFTlyze Dataset.csv")  # Replace with the actual path to your CSV file

# Step 2: Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Step 3: Plot the distribution of NFT sales over time
plt.figure(figsize=(12, 6))

# Group by month and calculate the sum of 'Sales_USD'
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Sales_USD'].sum()

# Plot the grouped data
monthly_sales.plot(kind='line', title='Monthly NFT Sales in USD')
plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.show()
```

```python
# 2. Characteristics of top-selling NFTs
# Analyzing distribution of 'Sales_USD'
plt.figure(figsize=(12, 6))
sns.histplot(df['Sales_USD'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of NFT Sales in USD')
plt.xlabel('Sales (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Correlation analysis between NFT attributes and sales performance
# Plotting the correlation heatmap
plt.figure(figsize=(10, 6))
correlation_matrix = df[['Sales_USD', 'Number_of_Sales', 'Active_Market_Wallets', 'Primary_Sales', 'Secondary_Sales_cumsum', 'AverageUSD_cum']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of NFT Attributes')
plt.show()
```

```python
# 4. Analyzing trends in cumulative data
plt.figure(figsize=(12, 6))
df[['Sales_USD_cumsum', 'Number_of_Sales_cumsum', 'Active_Market_Wallets_cumsum']].plot(title='Cumulative Trends in NFT Sales Metrics')
plt.xlabel('Date Index')
plt.ylabel('Cumulative Values')
plt.show()
```

```python
import pandas as pd

# Step 1: Load the dataset
df =  pd.read_csv(r"C:\Users\laksh\Downloads\NFTlyze Dataset.csv") # Replace with your dataset path

# Step 2: Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Step 3: Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Step 4: Resample the data to a monthly frequency
monthly_sales = df['Sales_USD'].resample('M').sum()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Step 1: Define and fit the ARIMA model
model = ARIMA(monthly_sales, order=(1, 1, 1))  # ARIMA(p, d, q) parameters
model_fit = model.fit()

# Step 2: Forecast future values
forecast_arima = model_fit.forecast(steps=12)  # Forecast for the next 12 months

# Step 3: Plot the results
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales, label='Actual Sales')
plt.plot(forecast_arima, label='Forecasted Sales (ARIMA)', linestyle='--', color='red')
plt.title('NFT Sales Forecast Using ARIMA')
plt.xlabel('Month')
plt.ylabel('Sales (USD)')
plt.legend()
plt.show()
```
